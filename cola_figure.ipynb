```python
import random

def overlap2(x):
  lim=10000-x
  s=[]
  s1=random.randint(0,lim)
  s2=random.randint(0,lim)
  if s1==s2:
    return 0
  s.append(s1)
  s.append(s2)
  s=sorted(s)
  #print(s)

  o=(s[0]+x)-s[1]
  if o<=0:
    o=0
  return (2*x-o)/10000

def overlap3(x):
  lim=10000-x
  s=[]
  s1=random.randint(0,lim)
  s2=random.randint(0,lim)
  s3=random.randint(0,lim)
  if s1==s2==s3:
    return 0
  s.append(s1)
  s.append(s2)
  s.append(s3)
  s=sorted(s)
  #print(s)

  o1=(s[0]+x)-s[1]
  o2=(s[1]+x)-s[2]
  o3=(s[0]+x)-s[2]
  if o1<=0:
    o1=0
  if o2<=0:
    o2=0
  #if o3<=0:
    #o3=0
  o=o1+o2
  return (3*x-o)/10000
```

```python
import random

def overlapo2(x):
  lim=10000-x
  s=[]
  s1=random.randint(0,lim)
  s2=random.randint(0,lim)
  if s1==s2:
    return 0
  s.append(s1)
  s.append(s2)
  s=sorted(s)
  #print(s)

  o=(s[0]+x)-s[1]
  if o<=0:
    o=0
  return (2*x-2*o)/(2*x)

def overlapo3(x):
  lim=10000-x
  s=[]
  s1=random.randint(0,lim)
  s2=random.randint(0,lim)
  s3=random.randint(0,lim)
  if s1==s2==s3:
    return 0
  s.append(s1)
  s.append(s2)
  s.append(s3)
  s=sorted(s)
  #print(s)

  o1=(s[0]+x)-s[1]
  o2=(s[1]+x)-s[2]
  o3=(s[0]+x)-s[2]
  if o1<=0:
    o1=0
  if o2<=0:
    o2=0
  if o3<=0:
    o3=0
  o=o1+o2
  #print(o)
  return (6*x-2*o1-2*o2-2*o3)/(6*x)
```
